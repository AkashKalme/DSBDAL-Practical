```python
X = df[["LSTAT", "RM"]]
y = df["MEDV"]
```

```python
print("X Shape: ", X.shape)
```

```python
print("y Shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print("X_train Shape: ", X_train.shape)
print("y_train Shape: ", y_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_test Shape: ", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
# Accuracy Metrics on Training Data
y_train_predict = linear_regressor.predict(X_train)

mae_train = mean_absolute_error(y_train, y_train_predict)
r2_train = r2_score(y_train, y_train_predict)
mse_train = mean_squared_error(y_train, y_train_predict)

print("Mean Absolute Error for Training Data: {}".format(round(mae_train, 2)))
print("R2 Score for Training Data: {}%".format(round(r2_train, 2)*100))
print("Mean Square Error for Training Data: {}".format(round(mse_train, 2)))
```

```python
# Accuracy Metrics on Test Data
y_test_predict = linear_regressor.predict(X_test)

mae_test = mean_absolute_error(y_test, y_test_predict)
r2_test = r2_score(y_test, y_test_predict)
mse_test = mean_squared_error(y_test, y_test_predict)

print("Mean Absolute Error for Test Data: {}".format(round(mae_test, 2)))
print("R2 Score for Test Data: {}%".format(round(r2_test, 2)*100))
print("Mean Squared Error for Test Data: {}".format(round(mse_test, 2)))
```
