Groupwise Statistical Analysis

```python
pd.set_option('display.max_columns', None)
df.groupby("Species").describe()
```

```python
df.groupby("Species").min()
```

```python
df.groupby("Species").max()
```

```python
df.groupby("Species").mean()
```

```python
df.groupby("Species").median()
```

```python
df.groupby("Species")[["SepalWidthCm","PetalLengthCm","PetalWidthCm"]].aggregate(pd.Series.mode)
```

```python
df.groupby("Species").sum()
```

```python
df.groupby("Species").std()
```

```python
df.groupby("Species").var()
```

```python
df.groupby("Species").cov()
```

```python
df.groupby("Species").quantile([0.25, 0.5, 0.75])
```

```python
df.groupby("Species").skew()
```
