4. Data Formatting and Data Normalization: Summarize the types of variables by checking the data types (i.e., character, numeric, integer, factor, and logical) of the variables in the data set. If variables are not in the correct data type, apply proper type conversions.

```python
df.head()
```

```python
df.dtypes
```

```python
# Date should be of the form DateTime
df["Date"] = df["Date"].astype("datetime64[ns]")
```

```python
df["Date"].dtype
```

```python
# Price, Bedroom, Bathroom, Car, Postcode, Landsize, YearBuilt, Propertycount should be of type int64
df["Price"] = df["Price"].astype("int64")
df["Bedroom2"] = df["Bedroom2"].astype("int64")
df["Bathroom"] = df["Bathroom"].astype("int64")
df["Postcode"] = df["Postcode"].astype("int64")
df["Landsize"] = df["Landsize"].astype("int64")
df["Propertycount"] = df["Propertycount"].astype("int64")
```

```python
df.dtypes
```

```python
df["Car"].unique()
```

```python
# Car contains NAN values fill them with 0
df['Car'] = df['Car'].fillna(0).astype(int)
```

```python
df["Car"].unique()
```

```python
df["Car"].dtype
```

```python
# Fill nan values of Building area with 0
df["BuildingArea"] = df["BuildingArea"].fillna(0)
```

```python
df["BuildingArea"].dtype
```

```python
# Change its datatype
df["BuildingArea"] = df["BuildingArea"].astype(int)
```

```python
df["BuildingArea"].dtype
```

```python
# Fill nan values of YearBuilt with 2000
df["YearBuilt"] = df["YearBuilt"].fillna(2000)
```

```python
df["YearBuilt"].dtype
```

```python
# Change its Datatype
df["YearBuilt"] = df["YearBuilt"].astype(int)
```

```python
df["YearBuilt"].dtype
```

```python
# Datatypes after conversion
df.dtypes
```

```python
# Get numeric column names
columns = list(df.columns)
numeric_columns = []
for column in columns:
    if(df[column].dtype in ['float', 'int', 'float64', 'int64', 'int32']):
        numeric_columns.append(column)
numeric_columns
```

```python
# Min-Max Normalization
min_max_normalized_df = df
for column in numeric_columns:
    min_max_normalized_df[column] = (min_max_normalized_df[column]-min_max_normalized_df[column].min())/(min_max_normalized_df[column].max()-min_max_normalized_df[column].min())
```

```python
min_max_normalized_df
```

```python
# Z-Score Normalization
z_score_normalized_df = df
for column in numeric_columns:
    z_score_normalized_df[column] = (z_score_normalized_df[column]-z_score_normalized_df[column].mean())/z_score_normalized_df[column].std()
```
